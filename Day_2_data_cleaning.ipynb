## Data Cleaning 1 - check missing or (abnormally high) zero values

```python
# Check for Missing Values. Is this easy to see? When to drop the lines and when to drop the column.
#isnull and isna are the same method, it's a stylistic preference
ds.isnull()
ds.isnull().sum()
```

```python
#difficult to read, filter out the columns with no nulls
#Get rid of all the columns with zero nulls
ds_null = ds.isnull().sum() ## gives you a series with the total number of null per column
ds_null

ds.isnull().sum()[ds.isnull().sum() > 0]
```

```python
#Check for zero values
ds_zeros = ds[ds == 0.0].count()
ds_zeros
ds_zeros[ds_zeros > 0]
```

```python
#Make a list to drop
ds_null = ds.isnull().sum()
ds_null[ds_null > 10000]
list(ds_null[ds_null > 10000].index)

ds_dropped = ds.drop(list(ds_null[ds_null > 10000].index), axis=1)
ds_dropped.head()
```

```python
## exercise: do the same for negative values
#.numeric()
# select_dtypes()
numeric = ds.select_dtypes(exclude=object)

ds_negatives = numeric[numeric < 0.0].count()
ds_negatives
ds_negatives[ds_negatives > 0]
```

```python
# Cylinders and displ have similar number of Null values. Maybe they have the same null entries?
# Let's check if it makes sense to drop them or keep them
null_displ = ds[ds['displ'].isnull()]

null_displ = null_displ[['year', 'make', 'model', 'trany', 'drive','fuelType','cylinders', 'displ']]
null_displ
#we seem to have found a strong correlation
```

```python
# In case we decide they are important even with little data,
# we might want to fill the missing values with a value we choose. In this case we fill with zero.
#fillna has multiple options and similar functions you may want to check out: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.fillna.html
ds[['displ', 'cylinders']] = ds[['displ', 'cylinders']].fillna(0)
ds[['year', 'make', 'model', 'trany', 'drive','fuelType','cylinders', 'displ']][29670:30000]
```

```python
# filling NaN values with 0 is often not correct. fillna needs to be done cautiously.
# Popular choices are to fill NaN values with the average or median of the remaining values of the same column
# Sometimes we can even create predictive models to fill in the missing values from other values of the same row

#These methods of filling NaNs are called *imputation* and are a very popular interview question
```

## Data Cleaning 2 - Data Type Correction

Sometimes you might want to change the type of a value

```python
ds['cylinders'] = ds['cylinders'].astype('int64')
ds['cylinders'].dtypes
```

## Data Cleaning 3 - Cleaning Text and Removing Special Symbols

```python
print(set(ds['trany']))
```

```python
ds['trany'] = ds['trany'].str.replace('-', '')
print(set(ds['trany']))
```

```python
#the same technique can be used to normalize the data
ds['trany'] = ds['trany'].str.replace('Automatic', 'Auto')
print(set(ds['trany']))

ds['trany'] = ds['trany'].str.replace("Auto\(", "Auto (")
print(set(ds['trany']))

#regex is often your friend here
```

## Data Cleaning 4 - Removing Duplicates

```python
# concatenating (put two dataframes together, like merge or join)
frames = [ds, ds]

ds  = pd.concat(frames)
```

```python
#drop duplicates removes all duplicated lines
before = len(ds)
display(before)
ds = ds.drop_duplicates()
ds
after = len(ds)
display(after)
print('Number of duplicate records dropped: ', str(before - after))

# you can also remove lines that have duplicates only on some columns
before = len(ds)
droppedsubset = ds.drop_duplicates(subset=['make', 'model', 'year', 'displ', 'cylinders',
                                           'trany', 'drive', 'VClass','fuelType','barrels08',
                                           'city08', 'highway08', 'comb08', 'co2TailpipeGpm', 'fuelCost08'])
after = len(droppedsubset)
print('Number of duplicate records dropped, considering only some columns: ', str(before - after))

# by default you retain the first copy of the dropped lines from each replication
before = len(ds)
losefirst = ds.drop_duplicates(subset=['make', 'model', 'year', 'displ', 'cylinders',
                                       'trany', 'drive', 'VClass','fuelType','barrels08',
                                       'city08', 'highway08', 'comb08', 'co2TailpipeGpm', 'fuelCost08'],keep=False)
after = len(losefirst)
print('Number of duplicate records dropped, if not keeping first instance in dataset: ', str(before - after))
```
